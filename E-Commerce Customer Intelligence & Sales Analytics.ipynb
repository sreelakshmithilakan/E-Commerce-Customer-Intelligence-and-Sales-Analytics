```python
#final customer segment column.
def segment_customer(row):

    if row['R_score'] >= 4 and row['M_score'] >= 4:
        return 'Champion'

    elif row['R_score'] >= 4 and row['M_score'] <= 3:
        return 'Potential Loyalist'

    elif row['R_score'] <= 2 and row['M_score'] >= 4:
        return 'At Risk'

    elif row['R_score'] <= 2 and row['M_score'] <= 2:
        return 'Lost Customer'

    else:
        return 'Regular Customer'
```

**Pareto Analysis**

```python

pareto = (
    product_sales_clean
    .groupby('product_category_name_english')['price']
    .sum()
    .sort_values(ascending=False)
    .reset_index()
)
pareto.columns = [ 'category','revenue']
```

```python
#Revenue Contribution %
pareto['revenue_pct'] = (
    pareto['revenue']
    / pareto['revenue'].sum()
) * 100
pareto.head()
```

```python
#Cumulative Revenue %

pareto['cumulative_pct'] = (
    pareto['revenue_pct']
    .cumsum()
)
pareto.head(15)
```

```python
print("Total Categories:", len(pareto))

print(
    "Categories contributing to first 80% revenue:",
    len(pareto[pareto['cumulative_pct'] <= 80])
)
```

```python
(16 / 71) * 100
```

```python
import plotly.graph_objects as go

pareto_top = pareto.head(20)

fig = go.Figure()

# Revenue Contribution Bars
fig.add_bar(
    x=pareto_top['category'],
    y=pareto_top['revenue_pct'],
    name='Revenue Contribution (%)'
)

# Cumulative Revenue Line
fig.add_scatter(
    x=pareto_top['category'],
    y=pareto_top['cumulative_pct'],
    mode='lines+markers',
    name='Cumulative Revenue (%)',
    yaxis='y2'
)

# 80% Reference Line
fig.add_hline(
    y=80,
    line_dash='dash',
    line_color='red',
    annotation_text='80% Threshold'
)

fig.update_layout(
    title='Pareto Analysis: Revenue Concentration by Product Category',
    title_x=0.5,
    plot_bgcolor='white',
    paper_bgcolor='white',

    xaxis=dict(
        title='Category',
        tickangle=-45,
        showgrid=False,
        showline=True,
        linecolor='black'
    ),

    yaxis=dict(
        title='Revenue Contribution (%)',
        showgrid=True,
        gridcolor='#E5E5E5',
        showline=True,
        linecolor='black'
    ),

    yaxis2=dict(
        title='Cumulative Revenue (%)',
        overlaying='y',
        side='right',
        range=[0,100]
    ),

    legend=dict(
        orientation='h',
        y=1.08
    )
)

fig.show()
```

Although the marketplace offers 71 product categories, only 16 categories account for approximately 80% of total revenue. This indicates that revenue generation is highly concentrated within a relatively small portion of the product portfolio. Strategic focus on these high-performing categories can significantly influence overall business performance, while lower-contributing categories may represent diversification opportunities or portfolio optimization candidates.

**Product Portfolio Matrix**

What role does each category play in the product portfolio?

```python
portfolio = (
    product_sales_clean
    .groupby('product_category_name_english')
    .agg(
        revenue=('price', 'sum'),
        items_sold=('product_id', 'count'),
        avg_price=('price', 'mean')
    )
    .reset_index()
)

portfolio.head()
```

```python
portfolio.describe()
```

```python
portfolio.sort_values(
    'revenue',
    ascending=False
).head(10)
```

```python
portfolio.sort_values(
    'avg_price',
    ascending=False
).head(10)
```

```python
price_cutoff = portfolio['avg_price'].median()
volume_cutoff = portfolio['items_sold'].median()

print("Median Price:", round(price_cutoff,2))
print("Median Volume:", round(volume_cutoff,2))
```

```python
def portfolio_segment(row):

    if row['avg_price'] >= price_cutoff and row['items_sold'] >= volume_cutoff:
        return 'Portfolio Star'

    elif row['avg_price'] < price_cutoff and row['items_sold'] >= volume_cutoff:
        return 'Traffic Driver'

    elif row['avg_price'] >= price_cutoff and row['items_sold'] < volume_cutoff:
        return 'Premium Niche'

    else:
        return 'Low Impact'


portfolio['portfolio_segment'] = (
    portfolio.apply(
        portfolio_segment,
        axis=1
    )
)

portfolio.head()
```

```python
portfolio['portfolio_segment'].value_counts()
```

```python
#Portfolio Stars
portfolio[
    portfolio['portfolio_segment'] == 'Portfolio Star'
].sort_values(
    'revenue',
    ascending=False
).head(10)
```

```python
#Traffic Drivers
portfolio[
    portfolio['portfolio_segment'] == 'Traffic Driver'
].sort_values(
    'items_sold',
    ascending=False
).head(10)
```

```python
#Premium Niches
portfolio[
    portfolio['portfolio_segment'] == 'Premium Niche'
].sort_values(
    'avg_price',
    ascending=False
).head(10)
```

```python
#Compare Review Scores
review_impact = (
    delivery_reviews
    .groupby('delivery_status')
    .agg(
        avg_review=('review_score', 'mean'),
        orders=('order_id', 'count')
    )
    .reset_index()
)

review_impact
```

Late deliveries are not merely operational failures; they have a direct and measurable impact on customer satisfaction. # **Revenue Concentration Analysis**

```python
#Customer Revenue Table
customer_revenue = (
    orders[['order_id', 'customer_id']]
    .merge(
        payments[['order_id', 'payment_value']],
        on='order_id'
    )
)
```

```python
customer_revenue = (
    customer_revenue
    .groupby('customer_id')
    .agg(
        revenue=('payment_value', 'sum')
    )
    .reset_index()
)
```

```python
#Pareto Logic
customer_revenue = (
    customer_revenue
    .sort_values(
        'revenue',
        ascending=False
    )
)
```

```python
customer_revenue['revenue_pct'] = (
    customer_revenue['revenue']
    /
    customer_revenue['revenue'].sum()
) * 100
```

```python
customer_revenue['cumulative_pct'] = (
    customer_revenue['revenue_pct']
    .cumsum()
)
```

```python
#Customers Needed for 80%
customers_for_80_pct = len(
    customer_revenue[
        customer_revenue['cumulative_pct'] <= 80
    ]
)

```
